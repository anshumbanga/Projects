```python
df.columns
# separating numeric columns
numeric_columns = df.describe().columns
```

```python
# seperating categorical columns

object_columns = df.describe(include='object').columns
```

```python
# trying chi squared test for feature selection

ct = pd.crosstab(df.gender,df.Churn,margins=True)
import scipy.stats as stats
stats.chi2_contingency(ct)[1]
# object_columns
```

```python
score = []

for i in df.columns:
  if i in object_columns:
    ct = pd.crosstab(df[i],df.Churn,margins=True)
    score.append(stats.chi2_contingency(ct)[1])
score
```

```python
feature_importance = pd.Series(score,index=object_columns,name='p-value')
plt.figure(figsize=(10,6))
sns.barplot(x=feature_importance.index,y=feature_importance.values,order=feature_importance.sort_values().index)
plt.title('Feature Importance')
plt.xticks(rotation=90)
plt.show()
```

We are going to take 14 features which have passed the chi-square test

```python
passed_features_chi = feature_importance.sort_values().index.drop(['PhoneService','gender'])
```

```python
numeric_columns
# we forgot seniorcitizen to consider it as categorical variable

test1 = pd.crosstab(df.SeniorCitizen,df.Churn,margins=True)
print(stats.chi2_contingency(test1)[1])
# it does have very low p-value which means it is also neccessary
1.2285843790939117**-33
```

```python
from scipy.stats import mannwhitneyu
```

```python
score_numeric = []

print('Monthly Charges')
yes = df[df['Churn'] == 'Yes']['MonthlyCharges']
no = df[df['Churn'] == 'No']['MonthlyCharges']
print('yes\n',yes.head())
print('No\n',no.head())

stat, p = mannwhitneyu(yes, no)
print('-'*100)
print(f"MonthlyCharges p-value: {p}")
score_numeric.append(p)

print('Total Charges')
yes=df[df['Churn']=='Yes']['TotalCharges']
no=df[df['Churn']=='No']['TotalCharges']
stat, p = mannwhitneyu(yes, no)
print('-'*100)
print(f"Total Charges p-value: {p}")


score_numeric.append(p)
```

```python
selected_features = list(passed_features_chi)
selected_features.extend(numeric_columns)
```

```python
df_feature_select = df.loc[:,selected_features]
```

```python
df_feature_select.head()
df_feature_select.shape
```

#### Now we are left with 18 columns out of 21 columns

```python
df_feature_select.to_csv('telecom_churn_selected_features.csv',index=False)
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
x= df_feature_select.iloc[:,1:]
y = df_feature_select.Churn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# size

print('training size\t',x_train.shape[0])
print('test size\t',x_test.shape[0])
```

```python
x_train.columns
num_col = list(x_train.describe().columns)
cat_col = list(set(x_train.columns)-set(num_col))
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('trf1',OneHotEncoder(sparse_output=False,drop='first'),cat_col),
        ('trf2',StandardScaler(),num_col)
    ]
)
transformer
```

```python
transformer.fit_transform(x_train)
transformer.get_feature_names_out()
x_train_transformed = pd.DataFrame(transformer.fit_transform(x_train),columns=transformer.get_feature_names_out())
```

```python
x_test_transformed = pd.DataFrame(transformer.transform(x_test),columns=transformer.get_feature_names_out())
```

```python
y_train = y_train.str.replace('Yes','1')
y_train = y_train.str.replace('No','0')
y_train = y_train.astype(int)

y_test = y_test.str.replace('Yes','1')
y_test = y_test.str.replace('No','0')
y_test = y_test.astype(int)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(64,input_shape=(28,),activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train_transformed,y_train,epochs=100,validation_split=0.2)
```

```python
print(y_test.head())
print(np.round(model.predict(x_test_transformed)))
```

```python
from sklearn.metrics import accuracy_score
y_pred = np.round(model.predict(x_test_transformed))
accuracy_score(y_test,y_pred)
```
